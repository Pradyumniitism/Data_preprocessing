##### 4.Handling Outliers

###### By Using Inter Quantile Range

```python
q1=diamond['depth'].quantile(0.25)
q3=diamond['depth'].quantile(0.75)
```

```python
iqr=q3-q1
iqr
```

```python
outliers1=diamond[(diamond['depth']<(q1-1.5*iqr)) | (diamond['depth']>(q3+1.5*iqr))]
```

```python
outliers1.shape
```

```python
diamond.shape
```

###### By Using Z_Score Method

```python
from scipy.stats import zscore  #Essential
```

```python
diamond['z_Score']=zscore(diamond['depth'])
```

```python
outliers2=diamond[diamond['z_Score'].abs()>3]
print(outliers2)
```

```python
outliers2.shape
```

```python
diamondss = diamond[~((diamond['depth'] < (q1 - 1.5 * iqr)) | (diamond['depth'] > (q3 + 1.5 * iqr)))]
```
